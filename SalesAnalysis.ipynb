```python
import pandas as pd
df=pd.read_csv('sales.csv')
df.head()
```

# the most selling Product based irrespective of their Price

```python
df[['Description','Quantity']].groupby('Description').sum().sort_values('Quantity',ascending=False)
```

# number of unique StockCode in the United Kingdom according to the given Sales data?

```python
df[['StockCode','Country']][df['Country']=='United Kingdom'].nunique()
```

# Invoice number and Customer ID of the highest amount of purchase based on the given Sales data?

```python
df['Amount']=df['Quantity']*df['Price']
```

```python
df.head(5)
```

```python
df.groupby(['Invoice','Customer ID'])['Amount'].sum().sort_values(ascending=False)
```
